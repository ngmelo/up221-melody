Using CalEnvironScreen 4.0 data, I aim to explore how the risk of lead exposure that low-income households with children face is distributed across LA County.

```python
## Opening geopandas library
import pandas as pd
import geopandas as gpd
```

```python
## Loading CalEnviroScreen data and saving as dataframe
CESdata = gpd.read_file('calenviroscreen40shpf2021shp.zip')
```

```python
## Renamining columns
CESdata.columns.to_list()
```

```python
## Selecting/Trimming for variables of interest
CESdatatrimmed = ['Tract','County','ApproxLoc','TotPop19','Pesticide','PesticideP','Lead', 'Lead_P','Poverty', 'HousBurd', 'PopChar', 'geometry']
CESdata[CESdatatrimmed].sample(5)
```

```python
## Saving trimmed dataframe
CESdata_trimmed = CESdata[CESdatatrimmed].copy()
CESdata_trimmed
```

```python
## Exploring California County variable
CESdata_trimmed['County'].value_counts()
```

```python
## Verifying range of values for Lead and Pesticides variables
print(CESdata_trimmed['Lead'].min())
print(CESdata_trimmed['Lead'].max())
print(CESdata_trimmed['Lead_P'].min())
print(CESdata_trimmed['Lead_P'].max())
print(CESdata_trimmed['Pesticide'].min())
print(CESdata_trimmed['Pesticide'].max())
print(CESdata_trimmed['PesticideP'].min())
print(CESdata_trimmed['PesticideP'].max())

### there seems to be missing values for Lead variable
```

```python
## Further trimming for select variables 
CESdatatrimmed2 = ['Tract','County','ApproxLoc','Lead','geometry']
CESdata_trimmed2 = CESdata_trimmed[CESdatatrimmed2].copy()
CESdata_trimmed2
```

```python
## Creating new variable that scales lead exposure qualitatively
CESdata_trimmed2['LeadExp']=''
CESdata_trimmed2
```

```python
## Creating scale values for new Lead Exposure variable
CESdata_trimmed2['LeadExp'] = '5: Very High'
CESdata_trimmed2.loc[CESdata_trimmed2['Lead'] < 70, 'LeadExp'] = '4: High'
CESdata_trimmed2.loc[CESdata_trimmed2['Lead'] < 50, 'LeadExp'] = '3: Moderate'
CESdata_trimmed2.loc[CESdata_trimmed2['Lead'] < 30, 'LeadExp'] = '2: Low'
CESdata_trimmed2.loc[CESdata_trimmed2['Lead'] < 10, 'LeadExp'] = '1: Very Low'
CESdata_trimmed2.sample(10)
```

```python
## Trimming to only include census tracts in LA County
CESdata_LAC = CESdata_trimmed2[CESdata_trimmed2.County == 'Los Angeles']
CESdata_LAC
```

```python
## First attempt to plot lead exposure data for LA County
CESdata_LAC.plot()
```

```python
## Customizing map to visually optimize lead exposure gradient using color schemes available in Colormaps
CESdata_LAC.plot(
    figsize = (20,12),
    column = 'LeadExp',
    cmap = 'YlOrRd',
    legend = True,
    legend_kwds={
               'loc': 'upper right',
               'bbox_to_anchor':(1,1)
            })
```

We can see that the areas of LA County where there is high risk for lead exposure in children living in low-income communities with older housing is concentrated in the southernmost and northeastern most portions of the County (in and around South and Southeast Los Angeles and east of Lancaster and Palmdale in the southwestern portions of the Mojave Desert, respectively).

# Prepping General Plan PDFs

Many of the plan PDFs are large enough that they needed to be split into separate files so that they could be uploaded into the repository (which has a 25 MB maximum file size for uploads). These files have been saved in a separate "General Plan Sections" subfolder. They have been (re)merged here into whole plan documents, before being moved into separate "General Plan", "General Plan A", and "General Plan B" subfolders, and subsequently passed through code to extract the text from them. All of the sections for each plan had to be merged in specific order. (They were organized and named accordingly to make this process easier).

For whatever reason, my device does not recognize the files unless I include the full absolute path. Moreover, out of caution, I am choosing to initialize and then close and then reinitialize the merger process to merge plan sections for each plan/city.

I am also choosing not the delete the individual section files once they have been merged, because I would like to preserve them for future element-specific analyses (e.g. Land Use elements,  Open Space and Conservation elements).

*NOTE: At least on my laptop, I was unable to merge the individual "Gardena" plan elements/sections (which remain in the "General Plan Sections" subfolder). I ultimately chose to exclude this plan (among a couple others) from the analysis.*
